Tokenizing

```python
import tiktoken

encoding = tiktoken.encoding_for_model("gpt-4.1-mini")

tokens = encoding.encode("Virtusa is a global IT services company specializing in digital business transformation, product engineering, and IT consulting for industries like banking, healthcare, and technology")
```

```python
for token_id in tokens:
    token_text = encoding.decode([token_id])
    print(f"{token_id} = {token_text}")
```
